```python
#importing the csv dataset into google collab
#csv dataset of TESLA INC. stock price values
#csv file downloaded from Kaggle (https://www.kaggle.com/datasets/varpit94/tesla-stock-data-updated-till-28jun2021)

import pandas as pd

#Read the CSV file into a DataFrame
df = pd.read_csv('/content/drive/My Drive/tesla_data.csv')

#Print the CSV file which contains dates and the respective opening stock  rate, highest, lowest, and closing stock  rate
print(df)

```

```python
# Starting the analysis and manipulation of the dataset

# Import the pandas library and alias it as pd for convenience
import pandas as pd

print(df.head())

# Display various descriptive statistics of the tesla stock prices
# Getting mean, standard deviation, minimum and maximum of each of the categories of stock prices
print(df.describe())

# Print the column names of the dataframe
print(df.columns)

# Print the data types of each column in the categories of stock prices
print(df.dtypes)
sorted_df = df.sort_values(by='Open', ascending=True)


```

```python
import matplotlib.pyplot as plt

# Plot a histogram of the 'High' column in the categories of stock prices to understand the variation of when the stock prices went highest
df['High'].plot.hist()

plt.ylabel('Frequency')
plt.xlabel('High')
# Display the 'High' plot
plt.show()


# Plot a histogram of the 'Low' column in the categories of stock prices to understand the variation of when the stock prices went lowest
df['Low'].plot.hist()

plt.ylabel('Frequency')
plt.xlabel('Low')
# Display the 'Low' plot
plt.show()
```

```python
#CORRELATION ANALYSIS of Tesla Stock Prices
import seaborn as sns

# Plot a heatmap of the correlation matrix of the dataframe stock prices.
# The 'annot=True' ensures that correlation values are displayed on the heatmap.
sns.heatmap(df.corr(), annot=True)
```

```python
# Using seaborn, plotting a boxplot for the 'High' column of stock prices of the dataframe of tesla

sns.boxplot(df['High'])
plt.title('Boxplot of High')
plt.show()
```

```python
# Plotting a boxplot for the 'High' column of stock prices of the dataframe of tesla
sns.boxplot(df['Low'])
plt.title('Boxplot of Low')
plt.show()
```

```python
#Plotting a boxplot for the 'Close' column of stock prices of the dataframe of tesla

sns.boxplot(df['Close'])
plt.title('Boxplot of Close')
plt.show()
```

```python
#market captalization of stock prices
df['Close'].plot(figsize=(10,6))
plt.title('Closing Prices Over Time')
plt.ylabel('Closing Price')
plt.xlabel('Date')  # Assuming the index of df is Date
plt.grid(True)
plt.show()
```

```python
sns.pairplot(df[['Open', 'Close', 'High', 'Low']])
plt.suptitle('Scatter Plot Matrix', y=1.02)
plt.show()
```

```python
df['Price_Change'] = df['Close'].diff()

# Plotting the change in stock price over time
df['Price_Change'].plot(figsize=(10,6))
plt.title('Change in Stock Price Over Time')
plt.ylabel('Change in Price')
plt.xlabel('Date')  # Assuming the index of df is Date
plt.grid(True)
plt.show()
```

```python
df['Cumulative_Price_Change'] = df['Price_Change'].cumsum()

df['Cumulative_Price_Change'].plot(figsize=(10,6))
plt.title('Cumulative Change in Stock Price Over Time')
plt.ylabel('Cumulative Change in Price')
plt.xlabel('Date')
plt.grid(True)
plt.show()
```
